# 🤖 Chatbot using Hugging Face Transformers (GPT-2)

## Objective
Build a conversational chatbot using GPT-2 model.

```python
# Import libraries
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Load GPT-2 model
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("GPT-2 Model loaded successfully!")
```

```python
def chatbot():
    print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

    chat_history = ""

    while True:
        user_input = input("You: ")

        if user_input.lower() in ["exit", "quit"]:
            print("Chatbot: Goodbye!")
            break

        # Better prompt format
        prompt = chat_history + f"User: {user_input}\nChatbot:"

        inputs = tokenizer.encode(prompt, return_tensors='pt').to(device)

        outputs = model.generate(
            inputs,
            max_length=inputs.shape[1] + 60,   # limit response size
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,
            top_k=50,
            top_p=0.9,
            temperature=0.7,
            no_repeat_ngram_size=2,   # 🔥 stops repetition
            repetition_penalty=1.5    # 🔥 reduces loops
        )

        full_output = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Extract only chatbot response
        response = full_output[len(prompt):].strip()

        # Stop at next User if generated
        response = response.split("User:")[0].strip()

        print("Chatbot:", response)

        # Update history (limit size)
        chat_history += f"User: {user_input}\nChatbot: {response}\n"

        # 🔥 Prevent memory overflow
        if len(chat_history) > 1000:
            chat_history = chat_history[-1000:]
```

```python
# Run chatbot
chatbot()
```
